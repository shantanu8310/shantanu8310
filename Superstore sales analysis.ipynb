```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')
```

```python
stores = pd.read_csv('stores.csv', index_col=False)
```

```python
stores.head()
```

```python
stores.rename(columns = {'Store ID ': 'Store_ID'}, inplace=True)
```

```python
stores.info()
```

```python
stores.describe()
```

### Bigger store has more Item availability?

```python
sns.jointplot(data=stores, x='Store_Area', y='Items_Available')
plt.show()
```

### Bigger store area more daily customers in the store?

```python
sns.jointplot(data=stores, x='Store_Area', y='Daily_Customer_Count')
plt.show()
```

### Bigger store area more sales?

```python
sns.jointplot(data=stores, x='Store_Area', y='Store_Sales')
plt.show()
```

### More Items Available increase in store sales?

```python
sns.jointplot(data=stores, x='Items_Available', y='Store_Sales')
plt.show()
```

### More Customer increased sales?

```python
sns.jointplot(data=stores, x='Daily_Customer_Count', y='Store_Sales')
plt.show()
```

### Findings:
- There is no evidence that a store bigger in area means there will be increase number of customer daily visiting the store.
- Bigger store area means there are more items available to be sold.

```python
stores_top_sales = stores.sort_values('Store_Sales', ascending=False)
stores_top_sales
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(stores.corr(), cmap='Greens', annot=True)
plt.show()
```

### There is strong correlation between store area and items available which was also seen in the jointplot.
